```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs

data = pd.read_csv("Downloads/student-mat.csv")
data.head(150)
```

```python
data.info()
```

```python
dfn=data[['traveltime','studytime']]
dfn.head()
```

```python
sbs.lineplot(dfn,dashes=True)
```

```python
sbs.lineplot(dfn['traveltime'],dashes=True)
```

```python
sbs.lineplot(dfn['studytime'],dashes=True)
```

```python
sbs.lineplot(x=dfn['traveltime'],y=dfn['studytime'],dashes=True)
```

```python
corelation=dfn.corr()
print(corelation)
```

```python
sbs.heatmap(corelation,cmap ="YlGnBu")
plt.show()
```

```python
covar=dfn.cov()
print(covar)
```

```python
sbs.heatmap(covar,cmap='YlGnBu')
plt.show()
```
